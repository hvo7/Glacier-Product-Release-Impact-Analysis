```python
import pandas as pd
import numpy as np
from pathlib import Path
import os

base_dir = Path(os.getcwd())

pd.set_option('display.max_rows', 10)

data_dir = base_dir / "Dataset"

os.chdir(data_dir)
```

## Handle duplicate rows

```python
accts = pd.read_csv("accounts.csv")
```

```python
# Show duplicate rows
accts.loc[accts.duplicated(keep=False),:].sort_values(accts.columns.to_list())
```

```python
# Remove the duplicate rows 
accts = accts.drop_duplicates()
```

```python
accts.to_csv(Path("accounts.csv"), index=False)
```
